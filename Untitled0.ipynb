Crear un análisis simple de las ventas de un pequeño negocio, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados con Matplotlib. Los estudiantes trabajarán con los siguientes datos de ventas mensuales: Datos de ventas mensuales (en miles) data = {“Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas mensuales
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

# Crear un DataFrame
df_ventas = pd.DataFrame(data)

# Mostrar el DataFrame
print("DataFrame de Ventas:")
display(df_ventas)

# Realizar operaciones con NumPy en la columna 'Ventas'
ventas_np = np.array(df_ventas['Ventas'])

venta_promedio = np.mean(ventas_np)
venta_maxima = np.max(ventas_np)
venta_minima = np.min(ventas_np)
suma_total_ventas = np.sum(ventas_np)

print(f"\nVenta promedio: {venta_promedio:.2f} mil")
print(f"Venta máxima: {venta_maxima} mil")
print(f"Venta mínima: {venta_minima} mil")
print(f"Suma total de ventas: {suma_total_ventas} mil")
```

### Visualización de las Ventas Mensuales

Vamos a crear un gráfico de barras para visualizar las ventas de cada mes y una línea para mostrar la tendencia general.


```python
plt.figure(figsize=(12, 6))
plt.bar(df_ventas['Mes'], df_ventas['Ventas'], color='skyblue', label='Ventas Mensuales')
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], color='red', marker='o', linestyle='--', label='Tendencia de Ventas')
plt.xlabel('Mes')
plt.ylabel('Ventas (en miles)')
plt.title('Análisis de Ventas Mensuales')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
